```python
#converting into categorical value
df["company"] = df["company"].astype('category')
df["company_codes"] = df["company"].cat.codes
company_codes = dict(enumerate(df['company'].cat.categories))
```

```python
#converting into categorical value
df["location"] = df["location"].astype('category')
df["location_codes"] = df["location"].cat.codes
loc_codes = dict(enumerate(df['location'].cat.categories))
```

```python
df= df[['yearsofexperience', 'yearsatcompany','company_codes', 'location_codes','totalyearlycompensation']]
```

```python
np.random.seed(2)

n = len(df)

n_val = int(0.2 * n)
n_test = int(0.2 * n)
n_train = n - (n_val + n_test)

idx = np.arange(n)
np.random.shuffle(idx)

df_shuffled = df.iloc[idx]

#creating train, validation and test datas
df_train = df_shuffled.iloc[:n_train].copy()
df_val = df_shuffled.iloc[n_train:n_train+n_val].copy()
df_test = df_shuffled.iloc[n_train+n_val:].copy()

y_train_orig = df_train.totalyearlycompensation.values
y_val_orig = df_val.totalyearlycompensation.values
y_test_orig = df_test.totalyearlycompensation.values

y_train = (df_train.totalyearlycompensation.values)
y_val = (df_val.totalyearlycompensation.values)
y_test = (df_test.totalyearlycompensation.values)

del df_train['totalyearlycompensation']
del df_val['totalyearlycompensation']
del df_test['totalyearlycompensation']
```

```python
linear_reg_model = tf.keras.Sequential([layers.Dense(units=1)])
```

```python
linear_reg_model.predict(df_train)
```

```python
linear_reg_model.layers[0].kernel
```

```python
linear_reg_model.compile(optimizer=tf.optimizers.SGD(learning_rate=0.001),loss='mean_absolute_error')
```

```python
#fit the model
linear_reg_model.fit(df_train,
          y_train,
          epochs=150,
          verbose=1)
```

```python
sal_pred = linear_reg_model.predict(df_test)
```

```python
sal_pred
```

```python
#calc RMSE value for. prediction
from keras import backend as K
def root_mean_squared_error(y_test, sal_pred):
        return K.sqrt(K.mean(K.square(sal_pred - y_test)))
```

```python
root_mean_squared_error(y_test, sal_pred)
```
